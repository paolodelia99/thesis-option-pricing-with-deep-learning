```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
cl_option_chain_path = 'data/clj22-options-american-options-exp-03_17_22-show-all-%futuresoptionsview%-daily-03-07-2022.csv'
```

```python
cl_op_chain = pd.read_csv(cl_option_chain_path)
cl_op_chain
```

```python
cl_op_chain['Type'] = cl_op_chain['Strike'].apply(lambda x: 'Call' if 'C' in x else 'Put')
cl_op_chain['Strike'] = cl_op_chain['Strike'].apply(lambda x: float(x.strip('C').strip('P')))
cl_op_chain.set_index('Strike', inplace=True)
```

```python
cl_op_chain
```

```python
cl_op_chain[cl_op_chain.Type == 'Call']['Open Interest'].plot(kind='hist')
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

cl_op_chain[cl_op_chain.Type == 'Call']['Open Interest'].plot(ax=ax, kind='bar')

plt.show()
```

```python
df_ = cl_op_chain[cl_op_chain.index < 130]
df_ = df_[df_.index > 100]
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

df_[df_.Type == 'Call']['Open Interest'].plot(ax=ax, kind='bar')

plt.show()
```
